## Preparing the images

```python
import cv2
import os

video_path = '/content/results.mp4'

output_dir = '/content/GFPGAN/inputs/tests'
os.makedirs(output_dir, exist_ok=True)


cap = cv2.VideoCapture(video_path)


frame_number = 0


while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break


    image_filename = os.path.join(output_dir, f'frame_{frame_number:04d}.jpg')
    cv2.imwrite(image_filename, frame)
    frame_number += 1


    print(f'Saved {image_filename}')


cap.release()
cv2.destroyAllWindows()

print('Video conversion completed.')
```

## Merging Images to a Video

```python
from moviepy.editor import ImageSequenceClip
import os

def images_to_video(image_folder, output_file, fps=24):
    """
    Merges images into a single video file in MP4 format.

    Args:
        image_folder (str): The directory containing the images.
        output_file (str): The output video file name (including .mp4 extension).
        fps (int): Frames per second for the video (default: 24).
    """

    images = [os.path.join(image_folder, img) for img in sorted(os.listdir(image_folder))
              if img.endswith(('.png', '.jpg', '.jpeg'))]


    video_clip = ImageSequenceClip(images, fps=fps)


    video_clip.write_videofile(output_file, codec='libx264')


images_to_video("/content/GFPGAN/results/restored_imgs", "/content/output_video_2.mp4", fps=30)
```

## Separate Audio with Input Video

```python
from moviepy.editor import VideoFileClip


video = VideoFileClip('/content/results.mp4')


audio = video.audio


audio.write_audiofile('/content/output_audio.mp3')


audio.close()
video.close()

print("Audio has been extracted and saved as 'output_audio.mp3'")
```

## Merge Audio with Output Video

```python
from moviepy.editor import VideoFileClip, AudioFileClip


video = VideoFileClip('/content/output_video_2.mp4')


audio = AudioFileClip('/content/output_audio.mp3')


video_with_audio = video.set_audio(audio)


video_with_audio.write_videofile('/content/output_video_3.mp4', codec='libx264', audio_codec='aac')


video.close()
audio.close()

print("Video has been merged with the audio and saved as 'output_video.mp4'")
```
